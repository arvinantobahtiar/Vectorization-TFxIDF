```python
import pandas as pd
import regex as re
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
def remove_digits(text):
    return re.sub(r'\d', '', text)

df['content_clean'] = df['content'].apply(remove_digits)

df['content_clean']
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('indonesian'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

df['content_clean'] = df['content_clean'].apply(remove_stopwords)

df['content_clean']
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
```

```python
text = df['content_clean']
```

```python
text
```

```python
tfidf = TfidfVectorizer(min_df=10)
tfidf_matrix = tfidf.fit_transform(text)
print(tfidf_matrix)
```

Penggunana parameter min_df yaitu pengambilan term/kata yang frekuensinya paling dikit dari nilai yang ditetapkan, disini saya gunakan 10 sebagai threshold minimalnya.
max_df-nya tidak saya tentukan.

# Interpretasi
Format "(i, j) x.y" mewakili bahwa dalam dokumen i, istilah j memiliki skor TF-IDF sebesar x.y.

(0, 145) 0.1897773559909465: Pada dokumen pertama (dokumen 0), istilah di indeks 145 memiliki skor TF-IDF sekitar 0.1897773559909465.
dan seterusnya untuk istilah lain dalam dokumen 0 dan dokumen berikutnya.

Output ini menunjukkan skor TF-IDF untuk semua istilah dalam semua dokumen.

Skor TF-IDF yang lebih tinggi menunjukkan bahwa sebuah istilah penting dalam dokumen tertentu, dan sering digunakan dalam  pengambilan informasi .

Kata 'android' bernilai 0.180469 didasarkan rumus TF IDF yaitu:
**TF-IDF(t, d) = TF(t, d) * IDF(t)**,
dengan

**TF(t, d) = (jumlah kemunculan kata t dalam dokumen d) / (jumlah kata total dalam dokumen d)**,
dan
**IDF(t) = log(N / (jumlah dokumen yang mengandung kata t))**

Besarnya nilai didasarkan kemunculan kata tersebut di dokumen tersebut dan di dokumen-dokumen lainnya dalam korpus.
